# Senado Federal
Durante a pandemia da SASR-COV-2, popular Covid 19, muito setores, público ou privada, se reinventaram os seus modos de trabalho, buscando no trabalho remoto um aliado para garantir o distanciamento social e consequentemente, a redução de algumas despesas. 

O objetivo desta análise de dados é verificar se essa redução pode ser percebida nas Cotas para Exercício da Atividade Parlamentar dos Senadores (CEAPS). 

Para que se possa se possa observar os efeitos das restrições causado pela pandemia, irei analisar os dados da CEAPS dos anos de 2018, pré pandemia, 2019 e 2020. Uma observação final, essa análise não tem caráter partidário ou ideológico, somente o caráter educativo.

```python
#Importando as bibliotecas do Panda, SeaBorn e MatLibPlot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Aprofundamento da análise
Precisaremos, de agora em diante, construir novos Data Frames, consolidando alguns dados que considero relavantes a esta análise.

```python
#Criando um novo DataFrame para armazenar valores que serão obtidos dos DataFrames anteriores
#Inincialmente o novo DataFrame terá somente duas colunas, ANO e VALOR_REEMBOLSADO
COLUNAS = [
    'ANO',
    'VALOR_REEMBOLSADO'
]

dfTotalValor = pd.DataFrame(columns=COLUNAS)
```

```python
#Obtendo a soma de todos os valores reembolsados dis anos de 2018, 2019 e 2020
#Armazenando em variáveis para depois popular no novo DataFrame
v_2018 = df2018['VALOR_REEMBOLSADO'].sum()
v_2019 = df2019['VALOR_REEMBOLSADO'].sum()
v_2020 = df2020['VALOR_REEMBOLSADO'].sum()
d = {'ANO': ['2018', '2019', '2020'], 'VALOR_REEMBOLSADO': [v_2018,v_2019,v_2020]}
dfTotalValor = pd.DataFrame(data=d)

#Conferindo se os valores foram armazenados corretamente
dfTotalValor.head()
```

```python
#Exibir um grafico de barras com os dados armazenados no novo DataFrame
#Será percecbido um decréscimo no total dos valores reembolsados, mas essa redução só se apresentar concretamente em 2020
sns.barplot(x='ANO', y='VALOR_REEMBOLSADO', data=dfTotalValor).set(title='Valor reembolsado vs Ano', xlabel='ANOS', ylabel='VALORES')
sns.set(font_scale=3, rc={"figure.figsize": (40, 10)})

#Ajustando as legendas
plt.show()
```

```python
#Transforma em números o que foi percebido no gráfico anterior, criei uma nova coluna no DataFrame e armazenei os novos valores
dfTotalValor["DIFERENÇA EM %"] = (dfTotalValor["VALOR_REEMBOLSADO"].diff()/dfTotalValor["VALOR_REEMBOLSADO"].shift(1))*100
dfTotalValor.head()
```

```python
#Exibir um grafico de barras com os dados armazenados no novo DataFrame
#Será percecbido um decréscimo no total dos valores reembolsados
#mas essa redução só se apresentar concretamente em 2020 como podemos agora representar
sns.barplot(y='ANO', x='DIFERENÇA EM %', data=dfTotalValor).set(title='Ano vs Percentual de economia dos Valor reembolsado em relação a 2018', ylabel='ANOS', xlabel='Percentual (%)')
sns.set(font_scale=3, rc={"figure.figsize": (40, 10)})

#Ajustando as legendas
plt.show()
```

# Conclusão

Dos sete TIPOS DE DESPESAS, que foram os que tiveram os maiores valores reembolsado, alguns se comportaram de forma análoga ao que se esperaria para um período em que vivemos, de pandemia, e consequentemente com uma redução de despesas, em sua grande maioria os trabalhos legislativos sendo realizados de forma remota, não somente pelos senadores, mas também por suas equipes de apoio.

Ranking da três despesas, que se destacaram, se posso me expressar assim, de forma negativa:

1º Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar - Obteve um acréscimo de mais de 12%, comparando o ano 2018 ao de 2020;

2º Divulgação da atividade parlamentar - Obteve um acréscimo de mais de 6%, comparando o ano 2019 ao de 2020;

3º Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos - Obteve um acréscimo de 0,15%, comparando o ano 2019 ao de 2020.

Todas as outras demais despesas tiveram decréscimos quando comparadas, ora com ano de 2018, pré pandemia, ou com 2019, ano de início da pandemia.

Voltando um pouco no início dessa análise de dados, foi verificado uma redução na quantidade de valores reembolsados em no ano de 2020 que resultou somente uma redução de pouco mais de 26% no valor em reais, reembolsado aos senadores.
